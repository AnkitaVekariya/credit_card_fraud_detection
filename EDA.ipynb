```python
correlation_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, 
            annot=False,  # Set to True to show correlation values
            cmap='coolwarm', 
            fmt=".2f",
            linewidths=0.5)
plt.title("Correlation Heatmap of Features", fontsize=16)
plt.show()
```

```python
# Correlation of features with 'Class'
class_corr = correlation_matrix['Class'].sort_values(ascending=False)
# print(class_corr)

# Plot the most correlated features with 'Class'
plt.figure(figsize=(8, 6))
sns.barplot(x=class_corr.index, y=class_corr.values, palette="viridis")
plt.title("Correlation of Features with Fraud (Class)", fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

```python
print(class_corr)
print("Min: ",class_corr.min())
print("Max: ",class_corr.max())
```

```python
class_counts = df['Class'].value_counts()
print("Class Distribution:")
print(class_counts)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.countplot(df, x="Class", palette='coolwarm')
plt.title("Class Distribution (0: Legitimate, 1: Fraudulent)", fontsize=14)
plt.xlabel("Class")
plt.ylabel("Frequency")
plt.show()

legitimate_percentage = (class_counts[0] / (class_counts[0] + class_counts[1])) * 100
fraud_percentage = (class_counts[1] / (class_counts[0] + class_counts[1])) * 100
print(f"Percentage of Legitimate Transactions: {legitimate_percentage:.2f}%")
print(f"Percentage of Fraudulent Transactions: {fraud_percentage:.2f}%")
```

```python
# Plot distributions of 'Time' and 'Amount'
plt.figure(figsize=(12, 6))

# Time
plt.subplot(1, 2, 1)
sns.histplot(df['Time'], bins=50, color='blue', kde=True)
plt.title("Distribution of Time")

# Amount
plt.subplot(1, 2, 2)
sns.histplot(df['Amount'], bins=50, color='green', kde=True)
plt.title("Distribution of Amount")

plt.show()
```

```python
# Boxplot for 'Amount'
plt.figure(figsize=(8, 4))
sns.boxplot(df['Amount'])
plt.title("Boxplot of Transaction Amount")
plt.show()
```

```python
# Plot distributions of V1-V28
plt.figure(figsize=(25, 25))
for i, col in enumerate(df.columns[1:29]):
    plt.subplot(7, 4, i+1)
    sns.histplot(df[col], bins=50, kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Class'] == 0]['Amount'], bins=50, kde=True, color='blue', label='Non-Fraud')
sns.histplot(df[df['Class'] == 1]['Amount'], bins=50, kde=True, color='red', label='Fraud')
plt.title('Transaction Amount: Fraud vs Non-Fraud')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Convert 'Time' to hours for better interpretation
df['Time_Hours'] = df['Time'] / 3600

# Distribution of 'Time'
plt.figure(figsize=(10, 6))
sns.histplot(df['Time_Hours'], bins=50, kde=True)
plt.title('Distribution of Transaction Time (in Hours)')
plt.xlabel('Time (Hours)')
plt.ylabel('Frequency')
plt.show()

# Fraud distribution over time
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Class'] == 1]['Time_Hours'], bins=50, kde=True, color='red')
plt.title('Fraudulent Transactions Over Time (in Hours)')
plt.xlabel('Time (Hours)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Class'] == 0]['Time_Hours'], bins=50, kde=True, color='blue', label='Non-Fraud')
sns.histplot(df[df['Class'] == 1]['Time_Hours'], bins=50, kde=True, color='red', label='Fraud')
plt.title('Transaction Time: Fraud vs Non-Fraud')
plt.xlabel('Time (Hours)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

##  Key Insights from EDA 

- **Class Imbalance**: The dataset is highly imbalanced, with fraud cases making up a tiny fraction of the total transactions.
    
- **Transaction Amount**: Fraudulent transactions tend to be smaller in amount compared to legitimate ones.
    
- **Time Distribution**: Fraudulent transactions are spread across time, but there may be specific time periods with higher fraud activity.
    
- **Correlation**: Some PCA features (e.g., **V14, V12, V10**) show significant **negative correlation** with fraud, while others (e.g., **V2, V4, V11**) show **positive correlation**.
